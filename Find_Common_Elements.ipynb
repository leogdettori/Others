```python

"""
Please, load those functions before you start!!!

Functions Created by Leo Dettori on 02.03.2021
"""

#Extracts Uniprot ID into a dictionary from Uniprot Multi-Fasta File
#In Function Format

def ID_multifasta(file_path):
    #Creates Dictionary for the results
    Uniprot = {'ID':[], 'Name':[]}
    
    #Loads the multi-FASTA file and counts the number of lines (for your reference)
    c = open(file_path)
    lines = c.readlines()
    total_lines = len(lines)
    #print('\n'+"Total number of lines in multiFASTA file is: "+str(total_lines)+'\n')
    c.close()


    #Starts to cycle trhough the lines
    for i in lines:
        #Searches for the FASTA Header
        if '>' in i:
            #Splits current line into a list of items stored into split_line
            split_line = i.split('|')
            #print(split_line)
            #Starts to store Information into the Uniprot ID Dictionary
            Uniprot['ID'].append(split_line[1])
            Uniprot['Name'].append(split_line[2])
            
    #Checks number of proteins
    proteins = len(Uniprot['ID'])

    #Prints Number of proteins in the multi-FASTA file
    print("There are "+str(proteins)+" proteins in the multi-FASTA file."+'\n')
    
    return Uniprot

#Compares 2 Lists and Checks for Common Elements
#In Cunction Format

def compare_2_lists(List_1,List_2):
    #Program creates results dictionary
    Results = {'Common':[]}
    
    #Program Starts to loop through Lists
    for i1 in List_1:
        #print(i1)
        for i2 in List_2:
            if i1 == i2:
                Results['Common'].append(i2)
        #print(i2)
    #Prints Results       
    print('Common = ['+ str(Results['Common']).replace(']', '').replace('[', '') +']'+'\n')   
    
    return Results['Common']
    
```
